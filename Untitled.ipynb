```python
face_clsfr=cv2.CascadeClassifier('haarcascade.xml')

source=cv2.VideoCapture(2)

labels_dict={0:'MASK',1:'NO MASK'}
color_dict={0:(0,255,0),1:(0,0,255)}
```

```python
img_array = cv2.imread('with_mask.jpg')
gray=cv2.cvtColor(img_array,cv2.COLOR_BGR2GRAY)           
resized = cv2.resize(gray , (100,100))
normalized = resized / 255
cv2.imshow('live',gray)
```

```python
plt.imshow(gray)
```

```python
# gray=cv2.cvtColor(resized,cv2.COLOR_BGR2GRAY)
faces=face_clsfr.detectMultiScale(gray) 
```

```python
for (x,y,w,h) in faces:
    detected=cv2.rectangle(gray,(x,y),(x+w,y+h),(255,0,0),2)
    roi = detected[y:y+h, x:x+w]
```
